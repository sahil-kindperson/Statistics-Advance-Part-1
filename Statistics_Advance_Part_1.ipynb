```python
#3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

def bernoulli_pdf(x, p):
    """
    Compute the PDF of a Bernoulli distribution.

    Parameters:
    - x (int): The outcome (0 or 1)
    - p (float): Probability of success (0 <= p <= 1)

    Returns:
    - float: Probability of observing outcome x
    """
    if x == 1:
        return p
    elif x == 0:
        return 1 - p
    else:
        return 0.0



p = 0.7


print("P(X=1):", bernoulli_pdf(1, p))
print("P(X=0):", bernoulli_pdf(0, p))
```

```python
#9. Write a Python function to calculate Z-scores from a dataset and plot them.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_and_plot_z_scores(data):
    """
    Calculate Z-scores from a dataset and plot a histogram with standard normal curve.

    Parameters:
    - data (array-like): Input dataset (numeric values)
    """
    data = np.array(data)
    mean = np.mean(data)
    std = np.std(data)

    z_scores = (data - mean) / std

    plt.figure(figsize=(8, 5))
    count, bins, _ = plt.hist(z_scores, bins=50, density=True, alpha=0.6, color='lightcoral', edgecolor='black', label='Z-score Histogram')

    x = np.linspace(-4, 4, 1000)
    pdf = norm.pdf(x, 0, 1)
    plt.plot(x, pdf, 'b-', lw=2, label='Standard Normal PDF')

    plt.title('Z-score Distribution with Standard Normal Overlay', fontsize=14)
    plt.xlabel('Z-score')
    plt.ylabel('Density')
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.legend()
    plt.tight_layout()
    plt.show()

    return z_scores


# Simulate some data
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=1000)

# Calculate and plot Z-scores
z_scores = calculate_and_plot_z_scores(data)
```

```python
#14. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(x, mean, std_dev):
    """
    Calculate Z-score for a given data point.
    """
    return (x - mean) / std_dev

def plot_z_score(z):
    """
    Plot the standard normal distribution and mark the Z-score.
    """
    x_vals = np.linspace(-4, 4, 1000)
    y_vals = norm.pdf(x_vals, 0, 1)

    plt.figure(figsize=(8, 5))
    plt.plot(x_vals, y_vals, label='Standard Normal PDF', color='navy')
    plt.fill_between(x_vals, y_vals, where=(x_vals <= z), color='lightcoral', alpha=0.5, label=f'Area ‚â§ Z = {z:.2f}')
    plt.axvline(z, color='red', linestyle='--', label=f'Z = {z:.2f}')

    plt.title('Z-score in Standard Normal Distribution', fontsize=14)
    plt.xlabel('Z-score')
    plt.ylabel('Probability Density')
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.legend()
    plt.tight_layout()
    plt.show()

x = 85
mean = 70
std_dev = 10

z_score = calculate_z_score(x, mean, std_dev)
print(f"Z-score for x = {x} (mean = {mean}, std = {std_dev}): {z_score:.2f}")

plot_z_score(z_score)
```

```python
#14. Implement hypothesis testing using Z-statistics for a sample dataset.

import numpy as np
from scipy.stats import norm

def z_test_one_sample(sample_data, population_mean, population_std, alpha=0.05):
    # Calculate sample statistics
    sample_mean = np.mean(sample_data)
    sample_size = len(sample_data)

    z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

    p_value = 2 * (1 - norm.cdf(abs(z_stat)))

    print(f"Sample Mean      = {sample_mean:.2f}")
    print(f"Z-Statistic      = {z_stat:.4f}")
    print(f"P-value          = {p_value:.4f}")
    print(f"Significance Œ±   = {alpha}")

    if p_value < alpha:
        print("Result: Reject the null hypothesis (H‚ÇÄ)")
    else:
        print("Result: Fail to reject the null hypothesis (H‚ÇÄ)")

sample = [102, 98, 101, 105, 100, 97, 99, 103, 96, 104]
z_test_one_sample(sample, population_mean=100, population_std=15, alpha=0.05)
```

```python
#15. Create a confidence interval for a dataset using Python and interpret the result.

import numpy as np
from scipy import stats

def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)

    t_crit = stats.t.ppf((1 + confidence) / 2, df=n-1)

    margin_of_error = t_crit * std_err

    lower = mean - margin_of_error
    upper = mean + margin_of_error

    print(f"Sample Mean        = {mean:.2f}")
    print(f"{int(confidence*100)}% Confidence Interval = ({lower:.2f}, {upper:.2f})")
    print("\nInterpretation:")
    print(f"We are {int(confidence*100)}% confident that the true population mean lies between {lower:.2f} and {upper:.2f}.")

sample_data = [102, 98, 101, 105, 100, 97, 99, 103, 96, 104]
confidence_interval(sample_data, confidence=0.95)
```

```python
#16. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.

import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Step 1: Generate data from a normal distribution
np.random.seed(42)  # for reproducibility
mu = 70       # true population mean
sigma = 10    # population standard deviation
n = 30        # sample size

sample_data = np.random.normal(loc=mu, scale=sigma, size=n)

# Step 2: Calculate confidence interval for the sample mean
sample_mean = np.mean(sample_data)
sample_se = stats.sem(sample_data)  # standard error of the mean
confidence = 0.95

# t-critical value for two-tailed test
t_crit = stats.t.ppf((1 + confidence) / 2, df=n-1)

# Margin of error
margin_of_error = t_crit * sample_se

# Confidence interval bounds
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

# Step 3: Print results
print(f"Sample Mean = {sample_mean:.2f}")
print(f"95% Confidence Interval = ({ci_lower:.2f}, {ci_upper:.2f})")
print(f"\nInterpretation: We are 95% confident that the true population mean lies between {ci_lower:.2f} and {ci_upper:.2f}.")

# Step 4: Optional ‚Äì plot the data and CI
plt.figure(figsize=(8, 4))
plt.errorbar(1, sample_mean, yerr=margin_of_error, fmt='o', capsize=5, color='darkblue')
plt.axhline(mu, color='green', linestyle='--', label=f'True Mean = {mu}')
plt.title('95% Confidence Interval for the Mean')
plt.ylabel('Value')
plt.xticks([])
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
#18. Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Set Poisson distribution parameter (average rate of events)
lambda_val = 4  # average number of events

# Generate x values (number of events)
x = np.arange(0, 15)

# Calculate CDF values
cdf = poisson.cdf(x, mu=lambda_val)

# Print a few example values
for i in range(len(x)):
    print(f"P(X ‚â§ {x[i]}) = {cdf[i]:.4f}")

# Plot the CDF
plt.figure(figsize=(8, 5))
plt.step(x, cdf, where='post', label=f'Poisson CDF (Œª = {lambda_val})', color='purple')
plt.title('CDF of a Poisson Distribution')
plt.xlabel('Number of Events (k)')
plt.ylabel('P(X ‚â§ k)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#20. Write a Python program to compare the standard deviations of two datasets and visualize the difference

import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate two datasets
np.random.seed(42)
data1 = np.random.normal(loc=50, scale=5, size=1000)   # Smaller spread
data2 = np.random.normal(loc=50, scale=15, size=1000)  # Larger spread

# Step 2: Calculate standard deviations
std1 = np.std(data1, ddof=1)
std2 = np.std(data2, ddof=1)

print(f"Standard Deviation of Dataset 1: {std1:.2f}")
print(f"Standard Deviation of Dataset 2: {std2:.2f}")

# Step 3: Visualize using a bar chart
plt.figure(figsize=(10, 4))

# Bar chart comparison
plt.subplot(1, 2, 1)
plt.bar(['Dataset 1', 'Dataset 2'], [std1, std2], color=['skyblue', 'salmon'])
plt.title('Standard Deviation Comparison')
plt.ylabel('Standard Deviation')

# Step 4: Visualize using histograms
plt.subplot(1, 2, 2)
plt.hist(data1, bins=30, alpha=0.6, label='Dataset 1 (œÉ ‚âà 5)', color='skyblue')
plt.hist(data2, bins=30, alpha=0.6, label='Dataset 2 (œÉ ‚âà 15)', color='salmon')
plt.title('Distribution of Two Datasets')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
#21. Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution

import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate data from a normal distribution
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=1000)

# Step 2: Calculate Range
data_range = np.max(data) - np.min(data)

# Step 3: Calculate IQR (Interquartile Range)
q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)
iqr = q3 - q1

# Step 4: Print results
print(f"Range: {data_range:.2f}")
print(f"Q1 (25th percentile): {q1:.2f}")
print(f"Q3 (75th percentile): {q3:.2f}")
print(f"Interquartile Range (IQR): {iqr:.2f}")

# Step 5: Optional - Visualize with a boxplot
plt.boxplot(data, vert=False)
plt.title('Boxplot of Normally Distributed Data')
plt.xlabel('Value')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
#22. Implement Z-score normalization on a dataset and visualize its transformation

import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate a synthetic dataset (e.g., normal distribution)
np.random.seed(42)
data = np.random.normal(loc=60, scale=15, size=1000)

# Step 2: Z-score normalization
mean = np.mean(data)
std = np.std(data)
z_scores = (data - mean) / std

# Step 3: Print summary
print(f"Original Mean: {mean:.2f}, Original Std Dev: {std:.2f}")
print(f"Z-Score Mean: {np.mean(z_scores):.2f}, Z-Score Std Dev: {np.std(z_scores):.2f}")

# Step 4: Visualize original and normalized data
plt.figure(figsize=(12, 5))

# Original Data Histogram
plt.subplot(1, 2, 1)
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title('Original Data Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Z-Score Normalized Histogram
plt.subplot(1, 2, 2)
plt.hist(z_scores, bins=30, color='salmon', edgecolor='black')
plt.title('Z-Score Normalized Distribution')
plt.xlabel('Z-score')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
#23. Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.


import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis

def analyze_normal_distribution_skew_kurtosis(mean=0, std_dev=1, size=1000, visualize=True):
    # Generate data from a normal distribution
    data = np.random.normal(loc=mean, scale=std_dev, size=size)

    # Calculate skewness and kurtosis
    skewness = skew(data)
    excess_kurtosis = kurtosis(data)  # By default, returns excess kurtosis (kurtosis - 3)

    print(f"Skewness: {skewness:.4f}")
    print(f"Kurtosis (Excess): {excess_kurtosis:.4f}")
    print(f"Kurtosis (Actual): {excess_kurtosis + 3:.4f}")

    # Optional: plot histogram
    if visualize:
        plt.hist(data, bins=30, edgecolor='black', color='skyblue', alpha=0.7)
        plt.title("Histogram of Normally Distributed Data")
        plt.xlabel("Value")
        plt.ylabel("Frequency")
        plt.grid(True, linestyle='--', alpha=0.6)
        plt.tight_layout()
        plt.show()

    return skewness, excess_kurtosis

# Example usage
analyze_normal_distribution_skew_kurtosis()
```
